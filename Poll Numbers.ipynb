```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
matplotlib.rcParams['figure.figsize'] = [15, 10]

df = pd.read_csv('polls.tsv', delimiter='\t')
df = df.replace('--', np.nan)
NAMES = [
    "Biden",
    "Sanders",
    "Warren",
    "Buttigieg",
    "Bloomberg",
    "Klobuchar",
    "Yang",
    "Booker",
    "Gabbard",
    "Steyer",
    "Castro",
    "Delaney",
    "Bennet",
]
df['start_ds'] = df['Date'].apply(lambda x: datetime.strptime(x.split('-')[0].strip() + '/19', "%m/%d/%y"))
df['end_ds'] = df['Date'].apply(lambda x: datetime.strptime(x.split('-')[1].strip() + '/19', "%m/%d/%y"))
plot_df = df[df['Poll'] == 'Politico/Morning Consult']

plot_df = plot_df.reset_index()
df = plot_df.to_csv('polls_formatted.tsv', sep='\t')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
for n in NAMES:
    mask = plot_df[n].notna()
    plt.plot(plot_df['start_ds'][mask], plot_df[n][mask].astype(np.int), label=n)

ax.legend()
ax.set_title('Democratic Polling Numbers')
axes = fig.gca()
axes.set_ylim([0,60])
axes.set_yticks([0, 10, 20, 30, 40, 50])
plt.show()
```
